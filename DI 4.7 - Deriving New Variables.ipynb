## If-Then Statement for price labeling

```python
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Creating Coloumn named Price_label

df['price_range'] = df.apply(price_label, axis=1)
```

```python
# Running Price_label Function

df['price_range'].value_counts(dropna=False)
```

```python
# Check Most expensive product in subset

df['prices'].max()
```

## If-Then Statement loc()

```python
# High range loc()

df_ords_prods_merged.loc[df_ords_prods_merged['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Mid range loc()

df_ords_prods_merged.loc[(df_ords_prods_merged['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
# Low range loc()

df_ords_prods_merged.loc[df_ords_prods_merged['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Loc() Count

df_ords_prods_merged['price_range_loc'].value_counts(dropna=False)
```

```python
# Setting up for new column for busiest hours of the day

result = []

for value in df_ords_prods_merged["order_hour_of_day"]:
  if value in [10,11,14,15,13,12,16,9]:
    result.append("Most orders")
  elif value in [17,8,18,19,20,7,21,22]:
    result.append("Average orders")
  else:
    result.append("Fewest orders")
```
